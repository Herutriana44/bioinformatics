```python
def huruf(string):
  only_alpha = ""

  ## looping through the string to find out alphabets
  for char in string:

  ## ord(chr) returns the ascii value
  ## CHECKING FOR UPPER CASE
    if ord(char) >= 65 and ord(char) <= 90:
      only_alpha += char
    ## checking for lower case
    elif ord(char) >= 97 and ord(char) <= 122:
      only_alpha += char

  ## printing the string which contains only alphabets
  return only_alpha
```

```python
dirbase = "/content/gdrive/MyDrive/Toxicity Dataset/"
```

```python
import pandas as pd

df = pd.read_csv(dirbase+'DatasetToxicityResult.csv')
```

```python
df['label1']
```

```python
chem = ""
for i in range(0,len(df)):
  chem += df['isosmiles'][i]
```

```python
iso = ''.join(set(chem))
len(iso)
```

```python
charChem = huruf(iso)
print(len(charChem))
print(charChem)
```

```python
print(list(charChem))
```

```python
def DecompChem(chem):
  cod = list(charChem)
  decomp = []
  for i in range(0,len(cod)):
    aa = chem.count(cod[i])
    decomp.append(aa)
  result = dict(zip(cod, decomp))
  return result
```

```python
dr = pd.DataFrame()
```

```python
dr['isosmiles'] = ""
```

```python
dr.head()
```

```python
codChem = list(charChem)
len(codChem)
```

```python
for j in range(0,len(codChem)):
  col = codChem[j]
  dr[col] = 0
```

```python
dr['label'] = ''
```

```python
dr.head()
```

```python
for k in range(0,len(df)):
  ROW = []
  smile = df['isosmiles'][k]
  ROW.append(smile)
  #dr['isosmiles'][k] = smile
  DecompIso = DecompChem(smile)
  #print(DecompIso)
  for o in codChem:
    oo = DecompIso[o]
    ROW.append(oo)
  ROW.append(df['label1'][k])
  #print(len(ROW))
  #print(ROW)
  dr.loc[len(dr.index)] = ROW
```
